```python

import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
DG = nx.DiGraph()
```

```python
DG.add_nodes_from("ABCD")
```

```python
nx.draw(DG,with_labels=True)
```

```python
pr=nx.pagerank(DG, alpha=0.85)
pr
```

```python
DG.add_weighted_edges_from([("A", "B", 1), ("B", "C", 1),("C","D",1),("D","A",1)]) 
nx.draw(DG,with_labels=True)
plt.show()
```
